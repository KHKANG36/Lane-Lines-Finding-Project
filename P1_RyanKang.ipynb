## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):#Color Conversion
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
def canny(img, low_threshold, high_threshold): #Canny Edge Detection
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size): #Gaussian Blurring
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    #defining a blank mask to start with
    mask = np.zeros_like(img)    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    #Empty List Define 
    #Slope(m)
    Left_Line_Slope=[]
    Right_Line_Slope=[]
    #Y-Intercept(b)
    Left_Line_Intercept=[]
    Right_Line_Intercept=[]
    #Y Values
    Left_Line_Y=[]
    Right_Line_Y=[]
    #Y_Max/Min
    Y_Max=img.shape[0]
    Y_Min=img.shape[0]
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            Slope=(y2-y1)/(x2-x1)
            
            #Separte Left/Right Line and Takes Values
            if Slope < -0.3:
                Left_Line_Slope.append(Slope) #Left Line 
                Left_Line_Y.append(y1)
                Left_Line_Y.append(y2)
                Left_Line_Intercept.append(y1-Slope*x1)
                
            
            if Slope > 0.3:
                Right_Line_Slope.append(Slope) #Right Line 
                Right_Line_Y.append(y1)
                Right_Line_Y.append(y2)
                Right_Line_Intercept.append(y1-Slope*x1)
                
            Y_Min=min(Y_Min, y1, y2) #Y minimum value 
            
    
    # Averaging and expolating of left line 
    Avg_Left_Line_Slope = np.mean(Left_Line_Slope)
    Avg_Intercept_Left = np.mean(Left_Line_Intercept)
    X_Min_Left_Line=int((Y_Max - Avg_Intercept_Left)/Avg_Left_Line_Slope)
    X_Max_Left_Line=int((Y_Min - Avg_Intercept_Left)/Avg_Left_Line_Slope)
    cv2.line(img, (X_Min_Left_Line, Y_Max),(X_Max_Left_Line, Y_Min),[0,255,0],10)
    
    # Averaging and expolating of right line 
    Avg_Right_Line_Slope = np.mean(Right_Line_Slope)
    Avg_Intercept_Right = np.mean(Right_Line_Intercept)
    X_Min_Right_Line=int((Y_Min-Avg_Intercept_Right)/Avg_Right_Line_Slope)
    X_Max_Right_Line=int((Y_Max-Avg_Intercept_Right)/Avg_Right_Line_Slope)
    cv2.line(img, (X_Min_Right_Line, Y_Min),(X_Max_Right_Line, Y_Max),[0,255,0],10)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

## Test Images

Build your pipeline to work on the images in the directory "test_images"  
**You should make sure your pipeline works well on these images before you try the videos.**

````python
import os
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
os.listdir("test_images/")

# Constants
kernel_size=5
low_threshold=60
high_threshold=180

rho=1
theta=np.pi/180
threshold=15
min_line_length=20
max_line_gap=15

def pipeline(image):
    
    grayscale_image=grayscale(image)
    
    smoothing=gaussian_blur(grayscale_image, kernel_size)
    
    edge_detect=canny(smoothing, low_threshold, high_threshold)
    
    height=image.shape[0]
    width=image.shape[1]
    
    vertices=np.array([[(50,height),(width/2-45,height/2+50),(width/2+45,height/2+50),(width-50,height)]],dtype=np.int32)
    
    regions=region_of_interest(edge_detect,vertices) 
    
    hough_image=hough_lines(regions, rho, theta, threshold, min_line_length, max_line_gap)
    
    return weighted_img(hough_image, image)


for test_image in os.listdir("test_images/"):
    img=mpimg.imread('test_images/'+test_image)
    plt.imshow(pipeline(img))
    plt.show()
````
